```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parmetros
k = 8.99e9  # Constante de Coulomb
q = 1e-9    # Carga puntual en Culombios

# Crear una rejilla de puntos en el espacio
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)

# Calcular el campo elctrico
r = np.sqrt(X**2 + Y**2)
Ex = k * q * X / (r**3)
Ey = k * q * Y / (r**3)

# Plotear el campo elctrico
plt.figure()
plt.quiver(X, Y, Ex, Ey)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Campo elctrico generado por una carga puntual')
plt.axis('equal')
plt.show()
```

```python


# Parmetros
k = 8.99e9  # Constante de Coulomb
q = 1e-9    # Carga puntual en Culombios

# Crear una rejilla de puntos en el espacio
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
z = np.linspace(-2, 2, 20)
X, Y, Z = np.meshgrid(x, y, z)

# Calcular el campo elctrico
r = np.sqrt(X**2 + Y**2 + Z**2)
Ex = k * q * X / (r**3)
Ey = k * q * Y / (r**3)
Ez = k * q * Z / (r**3)

# Plotear el campo elctrico en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, Ex, Ey, Ez, length=0.1, normalize=True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Campo elctrico en 3D generado por una carga puntual')
plt.show()
```
